```python
import numpy as np
import pandas as pd
from IPython.display import HTML, display
employee_data = {
    "employee_id": [101, 102, 103, 104, 105],
    "name": ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "department": ["Sales", "Sales", "Engineering", "Engineering", "HR"],
    "sales": [50000, 45000, 0, 0, 0],
    "hours_worked": [160, 150, 180, 170, 140],
    "satisfaction_score": [8.5, 7.0, 6.5, 9.0, 5.5]
}
df = pd.DataFrame(employee_data)
df["productivity"] = np.where(df["sales"] > 0, df["sales"] / df["hours_worked"], 0)
z_hours = np.abs((df["hours_worked"] - df["hours_worked"].mean()) / df["hours_worked"].std())
z_satisfaction = np.abs((df["satisfaction_score"] - df["satisfaction_score"].mean()) / df["satisfaction_score"].std())

outliers_hours = df[z_hours > 2]["name"].tolist()
outliers_satisfaction = df[z_satisfaction > 2]["name"].tolist()
badges = []
for _, row in df.iterrows():
    badge = []
    if row["sales"] == df["sales"].max():
        badge.append("Top Seller 🏆")
    if row["hours_worked"] == df["hours_worked"].max():
        badge.append("Most Dedicated ⏱️")
    if row["satisfaction_score"] == df["satisfaction_score"].max():
        badge.append("Happiest 😊")
    if row["satisfaction_score"] == df["satisfaction_score"].min():
        badge.append("Low Morale ⚠️")
    badges.append(", ".join(badge) if badge else "-")
df["badges"] = badges
insights = []
sales_df = df[df["department"] == "Sales"]
if not sales_df.empty:
    insights.append(f"<strong>Average Sales (Sales Dept):</strong> ${sales_df['sales'].mean():,.2f}")
    insights.append(f"<strong>Total Sales (Sales Dept):</strong> ${sales_df['sales'].sum():,.2f}")
insights.append(f"<strong>Average Hours Worked:</strong> {df['hours_worked'].mean():.1f} hours")
insights.append(f"<strong>Average Satisfaction Score:</strong> {df['satisfaction_score'].mean():.1f}/10")
insights.append(f"<strong>Top Productivity:</strong> {df.loc[df['productivity'].idxmax(), 'name']} (${df['productivity'].max():.2f}/hr)")
insights.append(f"<strong>Outliers in Hours Worked:</strong> {', '.join(outliers_hours) if outliers_hours else 'None'}")
insights.append(f"<strong>Outliers in Satisfaction Score:</strong> {', '.join(outliers_satisfaction) if outliers_satisfaction else 'None'}")
table_html = df.to_html(index=False, classes="data-table")

insights_html = "<br>".join(f"<p>{insight}</p>" for insight in insights)

html_report = f"""
<style>
    body {{
        font-family: Arial, sans-serif;
    }}
    .tab {{
        display: none;
    }}
    .tab-button {{
        background: #444;
        border: none;
        color: #fff;
        padding: 10px 20px;
        cursor: pointer;
        display: inline-block;
    }}
    .tab-button:hover {{
        background: #666;
    }}
    .tab.active {{
        display: block;
        margin-top: 20px;
    }}
    .data-table {{
        width: 100%;
        border-collapse: collapse;
        margin-top: 15px;
    }}
    .data-table th, .data-table td {{
        border: 1px solid #ccc;
        padding: 8px;
        text-align: left;
    }}
    .data-table th {{
        background-color: #ffccff;
    }}
    .data-table tr:nth-child(even) {{
        background-color: #f2f2f2;
    }}
</style>

<div>
    <button class="tab-button" onclick="openTab('data')">Employee Data</button>
    <button class="tab-button" onclick="openTab('insights')">Insights</button>

    <div id="data" class="tab active">
        <h2>Employee Data</h2>
        {table_html}
    </div>

    <div id="insights" class="tab">
        <h2>Insights</h2>
        {insights_html}
    </div>
</div>

<script>
    function openTab(id) {{
        var tabs = document.querySelectorAll('.tab');
        tabs.forEach(tab => tab.classList.remove('active'));
        document.getElementById(id).classList.add('active');
    }}
</script>
"""
display(HTML(html_report))
```

"Comprehensive Employee Performance Analysis with Graphical Insights"- using Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(42)
num_employees = 50
departments = ['Sales', 'Marketing', 'Engineering', 'HR', 'Finance']
positions = ['Junior', 'Mid', 'Senior', 'Manager']

data = {
    'EmployeeID': np.arange(1001, 1001 + num_employees),
    'Name': [f'Employee_{i}' for i in range(1, num_employees + 1)],
    'Department': np.random.choice(departments, num_employees),
    'Position': np.random.choice(positions, num_employees, p=[0.3, 0.4, 0.2, 0.1]),
    'Tenure': np.random.randint(1, 10, num_employees),
    'ProjectsCompleted': np.random.randint(1, 20, num_employees),
    'PerformanceScore': np.random.normal(70, 15, num_employees).round(1),
    'SatisfactionScore': np.random.randint(1, 10, num_employees)
}
data['PerformanceScore'] = np.clip(data['PerformanceScore'], 0, 100)
df = pd.DataFrame(data)
print("Employee Performance Data Overview:")
print(df.info())
print("\nFirst 5 rows:")
print(df.head())
print("\nPerformance Score Statistics:")
print(df['PerformanceScore'].describe())
print("\nDepartment-wise Performance:")
dept_stats = df.groupby('Department')['PerformanceScore'].agg(['mean', 'median', 'std', 'count'])
print(dept_stats)
print("\nPosition-wise Performance:")
position_stats = df.groupby('Position')['PerformanceScore'].agg(['mean', 'median', 'std', 'count'])
print(position_stats)
print("\nCorrelation between Tenure and Performance Score:")
print(df[['Tenure', 'PerformanceScore']].corr())
def categorize_performance(score):
    if score >= 90:
        return 'Excellent'
    elif score >= 80:
        return 'Good'
    elif score >= 70:
        return 'Average'
    elif score >= 60:
        return 'Below Average'
    else:
        return 'Poor'

df['PerformanceCategory'] = df['PerformanceScore'].apply(categorize_performance)
print("\nPerformance Category Distribution:")
print(df['PerformanceCategory'].value_counts())
top_performers = df.nlargest(5, 'PerformanceScore')
print("\nTop 5 Performers:")
print(top_performers[['Name', 'Department', 'Position', 'PerformanceScore']])
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
plt.hist(df['PerformanceScore'], bins=15, edgecolor='black')
plt.title('Performance Score Distribution')
plt.xlabel('Performance Score')
plt.ylabel('Count')
plt.subplot(2, 2, 2)
dept_stats['mean'].sort_values().plot(kind='bar')
plt.title('Average Performance by Department')
plt.ylabel('Average Performance Score')
plt.subplot(2, 2, 3)
position_stats['mean'].sort_values().plot(kind='bar')
plt.title('Average Performance by Position')
plt.ylabel('Average Performance Score')
plt.subplot(2, 2, 4)
plt.scatter(df['Tenure'], df['PerformanceScore'], alpha=0.6)
plt.title('Tenure vs Performance Score')
plt.xlabel('Tenure (years)')
plt.ylabel('Performance Score')
plt.tight_layout()
plt.show()
print("\nProjects Completed vs Performance:")
print(df[['ProjectsCompleted', 'PerformanceScore']].corr())
print("\nSatisfaction Score vs Performance:")
print(df[['SatisfactionScore', 'PerformanceScore']].corr())
df.to_csv('employee_performance_analysis.csv', index=False)
print("\nData saved to 'employee_performance_analysis.csv'")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML
np.random.seed(42)
num_employees = 50
departments = ['Sales', 'Marketing', 'Engineering', 'HR', 'Finance']
positions = ['Junior', 'Mid', 'Senior', 'Manager']

data = {
    'EmployeeID': np.arange(1001, 1001 + num_employees),
    'Name': [f'Employee_{i}' for i in range(1, num_employees + 1)],
    'Department': np.random.choice(departments, num_employees),
    'Position': np.random.choice(positions, num_employees, p=[0.3, 0.4, 0.2, 0.1]),
    'Tenure': np.random.randint(1, 10, num_employees),
    'ProjectsCompleted': np.random.randint(1, 20, num_employees),
    'PerformanceScore': np.random.normal(70, 15, num_employees).round(1),
    'SatisfactionScore': np.random.randint(1, 10, num_employees)
}

data['PerformanceScore'] = np.clip(data['PerformanceScore'], 0, 100)
df = pd.DataFrame(data)
def categorize_performance(score):
    if score >= 90:
        return 'Excellent'
    elif score >= 80:
        return 'Good'
    elif score >= 70:
        return 'Average'
    elif score >= 60:
        return 'Below Average'
    else:
        return 'Poor'

df['PerformanceCategory'] = df['PerformanceScore'].apply(categorize_performance)
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
plt.hist(df['PerformanceScore'], bins=15, edgecolor='black')
plt.title('Performance Score Distribution')
plt.xlabel('Performance Score')
plt.ylabel('Count')
plt.subplot(2, 2, 2)
df.groupby('Department')['PerformanceScore'].mean().sort_values().plot(kind='bar', color='skyblue')
plt.title('Average Performance by Department')
plt.ylabel('Avg Score')
plt.subplot(2, 2, 3)
df.groupby('Position')['PerformanceScore'].mean().sort_values().plot(kind='bar', color='orange')
plt.title('Average Performance by Position')
plt.ylabel('Avg Score')
plt.subplot(2, 2, 4)
plt.scatter(df['Tenure'], df['PerformanceScore'], alpha=0.6)
plt.title('Tenure vs Performance Score')
plt.xlabel('Tenure (years)')
plt.ylabel('Performance Score')

plt.tight_layout()
plt.show()
top_perf = df.loc[df['PerformanceScore'].idxmax()]
top_sat = df.loc[df['SatisfactionScore'].idxmax()]
top_proj = df.loc[df['ProjectsCompleted'].idxmax()]

iqr_tenure = df['Tenure'].quantile(0.75) - df['Tenure'].quantile(0.25)
low_tenure = df['Tenure'].quantile(0.25) - 1.5 * iqr_tenure
high_tenure = df['Tenure'].quantile(0.75) + 1.5 * iqr_tenure
outliers_tenure = df[(df['Tenure'] < low_tenure) | (df['Tenure'] > high_tenure)]

def bar_chart(label, values, scale=50, unit=""):
    max_val = max(values) if len(values) > 0 else 1
    bars = ""
    for name, val in zip(df["Name"], values):
        bar_len = int((val / max_val) * scale)
        bars += f"<strong>{label} - {name}:</strong> {'█' * bar_len} {val:.2f}{unit}<br>"
    return bars

perf_bars = bar_chart("Performance", df["PerformanceScore"], unit="/100")
sat_bars = bar_chart("Satisfaction", df["SatisfactionScore"], unit="/10")
proj_bars = bar_chart("Projects", df["ProjectsCompleted"], unit="")

html_code = f"""
<div style="background-color:#26264D;padding:20px;color:white;border-radius:10px;font-family:sans-serif">
  <h1 style="color:#FF33A6;">Employee Performance Dashboard</h1>

  <h2> Key Metrics</h2>
  <ul>
    <li><strong>Average Performance Score:</strong> {df['PerformanceScore'].mean():.2f}/100</li>
    <li><strong>Top Performer:</strong> {top_perf['Name']} ({top_perf['PerformanceScore']} pts)</li>
    <li><strong>Highest Satisfaction:</strong> {top_sat['Name']} ({top_sat['SatisfactionScore']}/10)</li>
    <li><strong>Most Projects Completed:</strong> {top_proj['Name']} ({top_proj['ProjectsCompleted']} projects)</li>
    <li><strong>Performance Category Counts:</strong> {df['PerformanceCategory'].value_counts().to_dict()}</li>
    <li><strong>Outliers in Tenure:</strong> {"None" if outliers_tenure.empty else ", ".join(outliers_tenure['Name'])}</li>
  </ul>

  <h2> Performance Visuals (Text-Based)</h2>
  <div style="font-family:monospace;">
    {perf_bars}<br><br>
    {sat_bars}<br><br>
    {proj_bars}
  </div>

  <h2>Full Dataset Snapshot</h2>
  {df.head(10).to_html(index=False, classes='data', header=True, border=0)}
</div>
"""
display(HTML(html_code))
```
